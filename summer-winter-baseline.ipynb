## Winter AQI - monthly

```python
"""
Produces: monthly_aq_winter_baseline.csv

One row per station x pollutant x year x month for winter months (November, December).
Aggregates avg_aqi, avg_mean_reading, peak_reading, n_days.
Tags season = Winter and zone by county.

Input file (place in same directory or update path below):
    all_pollutants_daily_clean_2023-2025.csv
"""

import pandas as pd

# Paths — update if your file lives elsewhere
INPUT_FILE = 'data/all_pollutants_daily_clean_2023-2025.csv'
OUTPUT     = 'data/monthly_aq_winter_baseline.csv'

WINTER_MONTHS = [11, 12]

# Zone assignment by county
ZONE_MAP = {
    'Bergen':   'Transit Corridor',
    'Hudson':   'Transit Corridor',
    'Passaic':  'Transit Corridor',
    'Atlantic': 'Jersey Shore',
    'Monmouth': 'Jersey Shore',
    'Ocean':    'Jersey Shore',
    'Essex':    'Newark Gateway',
}

def assign_zone(county):
    return ZONE_MAP.get(county, 'Statewide')

# Load and filter
print('Loading all_pollutants_daily_clean_2023-2025.csv...')
df = pd.read_csv(INPUT_FILE, low_memory=False)
df['Date Local'] = pd.to_datetime(df['Date Local'])
df['Month'] = df['Date Local'].dt.month

df = df[df['Month'].isin(WINTER_MONTHS)].copy()
print(f'Rows after winter filter: {len(df):,}')

# Aggregate to one row per station x pollutant x year x month
print('Aggregating...')
summary = df.groupby(
    ['Local Site Name', 'County Name', 'Latitude', 'Longitude',
     'pollutant', 'Year', 'Month'],
    as_index=False
).agg(
    avg_aqi=('daily_max_aqi', 'mean'),
    avg_mean_reading=('daily_mean_reading', 'mean'),
    peak_reading=('daily_1st_max', 'max'),
    n_days=('Date Local', 'nunique')
)

# Tag season and zone
summary['season'] = 'Winter'
summary['zone']   = summary['County Name'].apply(assign_zone)

# Column order
summary = summary[[
    'Local Site Name', 'County Name', 'Latitude', 'Longitude',
    'pollutant', 'Year', 'Month', 'season', 'zone',
    'avg_aqi', 'avg_mean_reading', 'peak_reading', 'n_days'
]]

# Save
summary.to_csv(OUTPUT, index=False)
print(f'\nDone. {len(summary):,} rows written to {OUTPUT}')
print(f'Stations:   {summary["Local Site Name"].nunique()}')
print(f'Pollutants: {list(summary["pollutant"].unique())}')
print(f'Years:      {sorted(summary["Year"].unique())}')
print(f'Months:     {sorted(summary["Month"].unique())}')
print(f'Zones:      {summary["zone"].value_counts().to_dict()}')
print(f'Nulls in avg_aqi: {summary["avg_aqi"].isna().sum()}')
print()
print(summary.head(10).to_string(index=False))
```

## Summer Traffic - monthly

```python
"""
Produces: monthly_traffic_summer_baseline.csv

One row per station x year x month for summer months (June, July, August)
across 2023, 2024, and 2025.
Aggregates monthly_adt (average daily vehicle count) and n_days.
Tags season = Summer and zone by county code.

Input files (all expected in data/Traffic/):
    NJ_JUN_2023 (TMAS).VOL
    NJ_JUL_2023 (TMAS).VOL
    NJ_AUG_2023 (TMAS).VOL
    NJ_JUN_2024 (TMAS).VOL
    NJ_JUL_2024 (TMAS).VOL
    NJ_AUG_2024 (TMAS).VOL
    NJ_JUN_2025 (TMAS).VOL
    NJ_JUL_2025 (TMAS).VOL
    NJ_AUG_2025 (TMAS).VOL
    NJ_2025 (TMAS).STA
"""

import os
import pandas as pd

# Paths
TRAFFIC_DIR = 'data/Traffic'
STA_FILE    = os.path.join(TRAFFIC_DIR, 'NJ_2025_(TMAS).STA')
OUTPUT      = 'data/monthly_traffic_summer_baseline.csv'

# Exact filenames — note 2025 uses title case month abbreviations
VOL_FILES = {
    (6,  2023): 'NJ_JUN_2023 (TMAS).VOL',
    (7,  2023): 'NJ_JUL_2023 (TMAS).VOL',
    (8,  2023): 'NJ_AUG_2023 (TMAS).VOL',
    (6,  2024): 'NJ_JUN_2024 (TMAS).VOL',
    (7,  2024): 'NJ_JUL_2024 (TMAS).VOL',
    (8,  2024): 'NJ_AUG_2024 (TMAS).VOL',
    (6,  2025): 'NJ_JUN_2025 (TMAS).VOL',
    (7,  2025): 'NJ_JUL_2025 (TMAS).VOL',
    (8,  2025): 'NJ_AUG_2025 (TMAS).VOL',
}

HOUR_COLS = [f'hour_{str(h).zfill(2)}' for h in range(24)]

# Zone assignment by county code
# County codes are FIPS county codes (last 3 digits of 5-digit FIPS)
# NJ FIPS = 34XXX
ZONE_MAP = {
    '003': 'Transit Corridor',   # Bergen
    '017': 'Transit Corridor',   # Hudson
    '031': 'Transit Corridor',   # Passaic
    '001': 'Jersey Shore',       # Atlantic
    '025': 'Jersey Shore',       # Monmouth
    '029': 'Jersey Shore',       # Ocean
    '013': 'Newark Gateway',     # Essex
}

def assign_zone(county_code):
    if pd.isna(county_code):
        return 'Statewide'
    return ZONE_MAP.get(str(int(county_code)).zfill(3), 'Statewide')

# Load STA metadata
print('Loading STA metadata...')
sta = pd.read_csv(STA_FILE, sep='|', low_memory=False)
sta.columns = sta.columns.str.lower().str.strip()
sta = sta.drop_duplicates(subset='station_id').reset_index(drop=True)
sta['station_id'] = sta['station_id'].astype(str).str.strip()
sta = sta[['station_id', 'latitude', 'longitude', 'station_location', 'county_code']]
print(f'STA stations loaded: {len(sta):,}')

# Load and process each VOL file
all_chunks = []

for (month, year), filename in sorted(VOL_FILES.items()):
    fpath = os.path.join(TRAFFIC_DIR, filename)

    if not os.path.exists(fpath):
        print(f'WARNING: File not found, skipping — {fpath}')
        continue

    print(f'Loading {filename}...')
    vol = pd.read_csv(fpath, sep='|', low_memory=False)
    vol.columns = vol.columns.str.lower().str.strip()

    # Build date column from year/month/day fields
    vol['date'] = pd.to_datetime({
        'year':  vol['year_record'],
        'month': vol['month_record'],
        'day':   vol['day_record']
    })

    # Coerce hour columns to numeric and sum for daily total
    for col in HOUR_COLS:
        vol[col] = pd.to_numeric(vol[col], errors='coerce')
    vol['daily_total'] = vol[HOUR_COLS].sum(axis=1)

    # Cast station_id to str before groupby
    vol['station_id'] = vol['station_id'].astype(str).str.strip()

    # Collapse lanes and directions — one row per station per date
    daily = (
        vol.groupby(['station_id', 'date'], as_index=False)
        .agg(daily_vehicle_count=('daily_total', 'sum'))
    )

    daily['year']  = year
    daily['month'] = month

    all_chunks.append(daily)

# Combine all months
print('Combining all months...')
traffic = pd.concat(all_chunks, ignore_index=True)

# Attach STA metadata
print('Attaching station metadata...')
traffic = traffic.merge(sta, on='station_id', how='left')

missing_coords = traffic['latitude'].isna().sum()
if missing_coords > 0:
    print(f'WARNING: {missing_coords} rows missing coordinates after STA join')

# Aggregate to monthly ADT per station
print('Aggregating to monthly ADT...')
summary = (
    traffic.groupby(['station_id', 'station_location', 'county_code',
                     'latitude', 'longitude', 'year', 'month'], as_index=False)
    .agg(
        monthly_adt=('daily_vehicle_count', 'mean'),
        n_days=('date', 'nunique')
    )
)

# Tag season and zone
summary['season'] = 'Summer'
summary['zone']   = summary['county_code'].apply(assign_zone)

# Round ADT to whole numbers
summary['monthly_adt'] = summary['monthly_adt'].round(0).astype(int)

# Column order
summary = summary[[
    'station_id', 'station_location', 'county_code',
    'latitude', 'longitude', 'year', 'month', 'season', 'zone',
    'monthly_adt', 'n_days'
]]

# Save
summary.to_csv(OUTPUT, index=False)
print(f'\nDone. {len(summary):,} rows written to {OUTPUT}')
print(f'Stations:    {summary["station_id"].nunique()}')
print(f'Years:       {sorted(summary["year"].unique())}')
print(f'Months:      {sorted(summary["month"].unique())}')
print(f'Zones:       {summary["zone"].value_counts().to_dict()}')
print(f'Missing lat: {summary["latitude"].isna().sum()}')
print()
print(summary.head(10).to_string(index=False))
```

## Summer Traffic - monthly

```python
"""
Produces: monthly_traffic_summer_baseline.csv

One row per station x year x month for summer months (June, July, August)
across 2023, 2024, and 2025.
Aggregates monthly_adt (average daily vehicle count) and n_days.
Tags season = Summer and zone by county code.

Input files (all expected in data/Traffic/):
    NJ_JUN_2023 (TMAS).VOL
    NJ_JUL_2023 (TMAS).VOL
    NJ_AUG_2023 (TMAS).VOL
    NJ_JUN_2024 (TMAS).VOL
    NJ_JUL_2024 (TMAS).VOL
    NJ_AUG_2024 (TMAS).VOL
    NJ_JUN_2025 (TMAS).VOL
    NJ_JUL_2025 (TMAS).VOL
    NJ_AUG_2025 (TMAS).VOL
    NJ_2025 (TMAS).STA
"""

import os
import pandas as pd

# Paths
TRAFFIC_DIR = 'data/Traffic'
STA_FILE    = os.path.join(TRAFFIC_DIR, 'NJ_2025_(TMAS).STA')
OUTPUT      = 'data/monthly_traffic_summer_baseline.csv'

# Exact filenames — note 2025 uses title case month abbreviations
VOL_FILES = {
    (6,  2023): 'NJ_JUN_2023 (TMAS).VOL',
    (7,  2023): 'NJ_JUL_2023 (TMAS).VOL',
    (8,  2023): 'NJ_AUG_2023 (TMAS).VOL',
    (6,  2024): 'NJ_JUN_2024 (TMAS).VOL',
    (7,  2024): 'NJ_JUL_2024 (TMAS).VOL',
    (8,  2024): 'NJ_AUG_2024 (TMAS).VOL',
    (6,  2025): 'NJ_JUN_2025 (TMAS).VOL',
    (7,  2025): 'NJ_JUL_2025 (TMAS).VOL',
    (8,  2025): 'NJ_AUG_2025 (TMAS).VOL',
}

HOUR_COLS = [f'hour_{str(h).zfill(2)}' for h in range(24)]

# Zone assignment by county code
# County codes are FIPS county codes (last 3 digits of 5-digit FIPS)
# NJ FIPS = 34XXX
ZONE_MAP = {
    '003': 'Transit Corridor',   # Bergen
    '017': 'Transit Corridor',   # Hudson
    '031': 'Transit Corridor',   # Passaic
    '001': 'Jersey Shore',       # Atlantic
    '025': 'Jersey Shore',       # Monmouth
    '029': 'Jersey Shore',       # Ocean
    '013': 'Newark Gateway',     # Essex
}

def assign_zone(county_code):
    if pd.isna(county_code):
        return 'Statewide'
    return ZONE_MAP.get(str(int(county_code)).zfill(3), 'Statewide')

# Load STA metadata
print('Loading STA metadata...')
sta = pd.read_csv(STA_FILE, sep='|', low_memory=False)
sta.columns = sta.columns.str.lower().str.strip()
sta = sta.drop_duplicates(subset='station_id').reset_index(drop=True)
sta['station_id'] = sta['station_id'].astype(str).str.strip()
sta = sta[['station_id', 'latitude', 'longitude', 'station_location', 'county_code']]
print(f'STA stations loaded: {len(sta):,}')

# Load and process each VOL file
all_chunks = []

for (month, year), filename in sorted(VOL_FILES.items()):
    fpath = os.path.join(TRAFFIC_DIR, filename)

    if not os.path.exists(fpath):
        print(f'WARNING: File not found, skipping — {fpath}')
        continue

    print(f'Loading {filename}...')
    vol = pd.read_csv(fpath, sep='|', low_memory=False)
    vol.columns = vol.columns.str.lower().str.strip()

    # Build date column from year/month/day fields
    vol['date'] = pd.to_datetime({
        'year':  vol['year_record'],
        'month': vol['month_record'],
        'day':   vol['day_record']
    })

    # Coerce hour columns to numeric and sum for daily total
    for col in HOUR_COLS:
        vol[col] = pd.to_numeric(vol[col], errors='coerce')
    vol['daily_total'] = vol[HOUR_COLS].sum(axis=1)

    # Cast station_id to str before groupby
    vol['station_id'] = vol['station_id'].astype(str).str.strip()

    # Collapse lanes and directions — one row per station per date
    daily = (
        vol.groupby(['station_id', 'date'], as_index=False)
        .agg(daily_vehicle_count=('daily_total', 'sum'))
    )

    daily['year']  = year
    daily['month'] = month

    all_chunks.append(daily)

# Combine all months
print('Combining all months...')
traffic = pd.concat(all_chunks, ignore_index=True)

# Attach STA metadata
print('Attaching station metadata...')
traffic = traffic.merge(sta, on='station_id', how='left')

missing_coords = traffic['latitude'].isna().sum()
if missing_coords > 0:
    print(f'WARNING: {missing_coords} rows missing coordinates after STA join')

# Aggregate to monthly ADT per station
print('Aggregating to monthly ADT...')
summary = (
    traffic.groupby(['station_id', 'station_location', 'county_code',
                     'latitude', 'longitude', 'year', 'month'], as_index=False)
    .agg(
        monthly_adt=('daily_vehicle_count', 'mean'),
        n_days=('date', 'nunique')
    )
)

# Tag season and zone
summary['season'] = 'Summer'
summary['zone']   = summary['county_code'].apply(assign_zone)

# Round ADT to whole numbers
summary['monthly_adt'] = summary['monthly_adt'].round(0).astype(int)

# Column order
summary = summary[[
    'station_id', 'station_location', 'county_code',
    'latitude', 'longitude', 'year', 'month', 'season', 'zone',
    'monthly_adt', 'n_days'
]]

# Save
summary.to_csv(OUTPUT, index=False)
print(f'\nDone. {len(summary):,} rows written to {OUTPUT}')
print(f'Stations:    {summary["station_id"].nunique()}')
print(f'Years:       {sorted(summary["year"].unique())}')
print(f'Months:      {sorted(summary["month"].unique())}')
print(f'Zones:       {summary["zone"].value_counts().to_dict()}')
print(f'Missing lat: {summary["latitude"].isna().sum()}')
print()
print(summary.head(10).to_string(index=False))
```

## Summer & Winter Comparison - aqi

```python
"""
Produces: county_aqi_seasonal_comparison.csv

Compares summer vs. winter AQI at the county level for each pollutant.
One row per County x Pollutant x Month showing summer avg_aqi, winter avg_aqi,
the delta, and the pct change from winter to summer.

Winter avg_aqi is averaged across all winter months (Nov + Dec) per county
per pollutant as the baseline reference for each summer month comparison.

Input files (place in same directory or update paths below):
    data/monthly_aq_summer_baseline.csv
    data/monthly_aq_winter_baseline.csv
"""

import pandas as pd

# Paths
SUMMER_FILE = 'data/monthly_aq_summer_baseline.csv'
WINTER_FILE = 'data/monthly_aq_winter_baseline.csv'
OUTPUT      = 'data/county_aqi_seasonal_comparison.csv'

# Load
print('Loading AQ baselines...')
summer = pd.read_csv(SUMMER_FILE)
winter = pd.read_csv(WINTER_FILE)

# Aggregate summer to county x pollutant x year x month
# Average across all stations in the county for that month
summer_county = summer.groupby(
    ['County Name', 'pollutant', 'zone', 'Year', 'Month'],
    as_index=False
).agg(
    summer_avg_aqi=('avg_aqi', 'mean'),
    summer_peak_aqi=('peak_reading', 'max'),
    summer_n_stations=('Local Site Name', 'nunique')
)

# Aggregate winter to county x pollutant baseline
# Average across all winter months and years — this is the single
# winter reference value each summer month is compared against
winter_county = winter.groupby(
    ['County Name', 'pollutant'],
    as_index=False
).agg(
    winter_avg_aqi=('avg_aqi', 'mean'),
    winter_peak_aqi=('peak_reading', 'max'),
    winter_n_stations=('Local Site Name', 'nunique')
)

# Join summer rows to winter baseline on County + Pollutant
print('Joining summer to winter baseline...')
comparison = summer_county.merge(
    winter_county[['County Name', 'pollutant', 'winter_avg_aqi', 'winter_peak_aqi']],
    on=['County Name', 'pollutant'],
    how='left'
)

# Compute delta and pct change
comparison['aqi_delta']      = (comparison['summer_avg_aqi'] - comparison['winter_avg_aqi']).round(2)
comparison['aqi_pct_change'] = ((comparison['aqi_delta'] / comparison['winter_avg_aqi']) * 100).round(1)

# Column order
comparison = comparison[[
    'County Name', 'zone', 'pollutant', 'Year', 'Month',
    'summer_avg_aqi', 'winter_avg_aqi', 'aqi_delta', 'aqi_pct_change',
    'summer_peak_aqi', 'winter_peak_aqi', 'summer_n_stations'
]].sort_values(['County Name', 'pollutant', 'Year', 'Month']).reset_index(drop=True)

# Round AQI columns
for col in ['summer_avg_aqi', 'winter_avg_aqi', 'summer_peak_aqi', 'winter_peak_aqi']:
    comparison[col] = comparison[col].round(1)

# Save
comparison.to_csv(OUTPUT, index=False)
print(f'\nDone. {len(comparison):,} rows written to {OUTPUT}')
print(f'Counties:   {comparison["County Name"].nunique()}')
print(f'Pollutants: {list(comparison["pollutant"].unique())}')
print(f'Years:      {sorted(comparison["Year"].unique())}')
print(f'Months:     {sorted(comparison["Month"].unique())}')
print()

# Shore counties ozone pct change
print('--- Jersey Shore ozone summer vs winter ---')
shore_ozone = comparison[
    (comparison['zone'] == 'Jersey Shore') &
    (comparison['pollutant'] == 'Ozone')
][['County Name', 'Year', 'Month', 'summer_avg_aqi', 'winter_avg_aqi', 'aqi_pct_change']]
print(shore_ozone.to_string(index=False))
```

## Summer & Winter Comparison - traffic

```python
"""
Produces: county_traffic_seasonal_comparison.csv

Compares summer vs. winter average daily traffic (ADT) at the county level.
One row per County x Year x Month showing summer monthly_adt, winter monthly_adt,
the delta, and the pct change from winter to summer.

Winter ADT is averaged across all winter months (Nov + Dec) per county
as the baseline reference for each summer month comparison.

Input files (place in same directory or update paths below):
    data/monthly_traffic_summer_baseline.csv
    data/monthly_traffic_winter_baseline.csv
"""

import pandas as pd

# Paths
SUMMER_FILE = 'data/monthly_traffic_summer_baseline.csv'
WINTER_FILE = 'data/monthly_traffic_winter_baseline.csv'
OUTPUT      = 'data/county_traffic_seasonal_comparison.csv'

# County code to name mapping (NJ FIPS county codes)
COUNTY_NAMES = {
    1:  'Atlantic',   2:  'Unknown',    3:  'Bergen',
    5:  'Burlington', 7:  'Camden',     9:  'Cape May',
    11: 'Cumberland', 13: 'Essex',      15: 'Gloucester',
    17: 'Hudson',     19: 'Hunterdon',  21: 'Mercer',
    23: 'Middlesex',  25: 'Monmouth',   27: 'Morris',
    29: 'Ocean',      31: 'Passaic',    33: 'Salem',
    35: 'Somerset',   37: 'Sussex',     39: 'Union',
    41: 'Warren',
}

# Load
print('Loading traffic baselines...')
summer = pd.read_csv(SUMMER_FILE)
winter = pd.read_csv(WINTER_FILE)

# Map county_code to county_name in both files
for df in [summer, winter]:
    df['county_name'] = df['county_code'].apply(
        lambda x: COUNTY_NAMES.get(int(x), 'Unknown') if not pd.isna(x) else 'Unknown'
    )

# Aggregate summer to county x year x month
# Average ADT across all stations in the county for that month
summer_county = summer.groupby(
    ['county_name', 'zone', 'year', 'month'],
    as_index=False
).agg(
    summer_monthly_adt=('monthly_adt', 'mean'),
    summer_n_stations=('station_id', 'nunique')
)

# Aggregate winter to county baseline
# Average across all winter months and years — single reference value
winter_county = winter.groupby(
    ['county_name'],
    as_index=False
).agg(
    winter_monthly_adt=('monthly_adt', 'mean'),
    winter_n_stations=('station_id', 'nunique')
)

# Join summer to winter baseline on county_name
print('Joining summer to winter baseline...')
comparison = summer_county.merge(
    winter_county[['county_name', 'winter_monthly_adt']],
    on='county_name',
    how='left'
)

# Compute delta and pct change
comparison['adt_delta']      = (comparison['summer_monthly_adt'] - comparison['winter_monthly_adt']).round(0)
comparison['adt_pct_change'] = ((comparison['adt_delta'] / comparison['winter_monthly_adt']) * 100).round(1)

# Round ADT columns
comparison['summer_monthly_adt'] = comparison['summer_monthly_adt'].round(0)
comparison['winter_monthly_adt'] = comparison['winter_monthly_adt'].round(0)

# Column order
comparison = comparison[[
    'county_name', 'zone', 'year', 'month',
    'summer_monthly_adt', 'winter_monthly_adt',
    'adt_delta', 'adt_pct_change',
    'summer_n_stations'
]].sort_values(['county_name', 'year', 'month']).reset_index(drop=True)

# Save
comparison.to_csv(OUTPUT, index=False)
print(f'\nDone. {len(comparison):,} rows written to {OUTPUT}')
print(f'Counties:  {comparison["county_name"].nunique()}')
print(f'Years:     {sorted(comparison["year"].unique())}')
print(f'Months:    {sorted(comparison["month"].unique())}')
print(f'Zones:     {comparison["zone"].value_counts().to_dict()}')
print()

# Shore counties pct change
print('--- Jersey Shore traffic summer vs winter ---')
shore = comparison[comparison['zone'] == 'Jersey Shore'][
    ['county_name', 'year', 'month', 'summer_monthly_adt', 'winter_monthly_adt', 'adt_pct_change']
]
print(shore.to_string(index=False))
```
